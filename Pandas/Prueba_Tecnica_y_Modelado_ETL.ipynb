## Parte Financiera

```python
finanzas = invoices_products.copy()
```

```python
finanzas = finanzas.merge(invoices_header.copy(), on = "Invoice")
```

```python
finanzas = finanzas.merge(suppliers.copy(), how = 'left',left_on="Supplier", right_on= "IDSupplier")
```

```python
finanzas.drop(columns=["Invoice","SupplierName","PaymentMethod","Product","Supplier","IDSupplier","OrderDate","InboundDate","PaymentTerms","Country"],inplace = True)
```

```python
finanzas
```

```python
finanzas["Total_Pedido"] = finanzas.Quantity.multiply(finanzas.Purchase_Price_Unit)
```

```python
finanzas
```

```python
finanzas = finanzas.drop(columns=["Quantity","Purchase_Price_Unit"]).sort_values(by=["InvoiceDate","Section"],ascending=True)
```

```python
finanzas
```

```python
finanzas=finanzas.merge(daily_currencies, how="left", left_on=["InvoiceDate","Currency"], right_on=["Date","Currency"], ).drop(columns="Date")
finanzas.fillna(1,inplace=True)
finanzas
```

```python
finanzas.Total_Pedido = finanzas.Total_Pedido.divide(finanzas.Close)
```

```python
finanzas = finanzas.drop(columns=["Open","Close","High","Low","Currency"]).rename(columns={"Section":"Seccion","InvoiceDate":"Fecha"})
```

```python
finanzas
```

```python
finanzas = finanzas.groupby(by=["Fecha","Seccion"]).Total_Pedido.sum().unstack().fillna(0)
```

```python
finanzas = finanzas.sort_values(by="Fecha",ascending=True).round(2)
```

```python
finanzas
```

## Calidad Del Proveedor

```python
ranking_proveedores = invoices_header.copy()
```

```python
ranking_proveedores["Tiempo_Entrega"] = (ranking_proveedores.InvoiceDate - ranking_proveedores.OrderDate)
```

```python
ranking_proveedores=ranking_proveedores.merge(suppliers.copy(), how= "left", left_on="Supplier" , right_on = "IDSupplier")
```

```python
ranking_proveedores = ranking_proveedores.merge(invoices_products.copy(), how="inner", on="Invoice")
```

```python
ranking_proveedores.Country.value_counts()
```

```python
in_spain = ranking_proveedores.Country == "ES"
in_europe = ranking_proveedores.Country.isin(["CH","IE","FR","DE"])
```

```python
ranking_proveedores["Variado_Envio"] = ranking_proveedores.Tiempo_Entrega.dt.days - 45
ranking_proveedores.loc[in_spain,"Variado_Envio"] = ranking_proveedores.Tiempo_Entrega.dt.days - 10
ranking_proveedores.loc[in_europe,"Variado_Envio"] = ranking_proveedores.Tiempo_Entrega.dt.days - 20
```

```python
ranking_proveedores
```

```python
ranking_proveedores.drop(columns=["Supplier","SupplierName","Invoice","InboundDate","InvoiceDate","Tiempo_Entrega","PaymentMethod","PaymentTerms","Currency","Quantity","Purchase_Price_Unit","Section"],inplace=True)
```

```python
ranking_proveedores
```

```python
ranking_proveedores.OrderDate = ranking_proveedores.OrderDate.dt.year
ranking_proveedores.rename(columns={"OrderDate":"Año"},inplace=True)
```

```python
ranking_proveedores=ranking_proveedores.groupby(by = ["Año","Country","Product","IDSupplier"]).Variado_Envio.mean().unstack()
```

```python
ranking_proveedores
```

Sin melt

```python
ranking_no_melt = ranking_proveedores.copy().reset_index()
```

```python
ranking_no_melt=ranking_no_melt.set_index(keys=["Año","Country","Product"]).sort_values(by=["Año","Country"], ascending=True, na_position="last")
```

```python
ranking_no_melt
```

Con melt

```python
ranking_proveedores=ranking_proveedores.melt(ignore_index=False).dropna().rename(columns = {"value":"Media_Envios"}).reset_index(drop=False)
```

```python
ranking_proveedores
```

```python
in_europe = ranking_proveedores.Country.isin(["CH","IE","FR","DE"])
```

```python
ranking_proveedores["Zona"] = ranking_proveedores.Country
ranking_proveedores.loc[in_europe,"Zona"] = "EUR"
ranking_proveedores = ranking_proveedores.drop(columns="Country")
```

```python
ranking_proveedores=ranking_proveedores.set_index(keys=["Año","IDSupplier","Zona","Product"])
```

La media marca los dias que ha tardado el pedido en llegar respecto de lo que se esperaba, los numeros negativos marcan tantos dias antes respecto de la fecha esperada

```python
ranking_proveedores.sort_values(by = ["Año","Media_Envios","IDSupplier","Zona","Product"], ascending=True)
```
